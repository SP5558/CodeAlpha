```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

print("✅ Libraries loaded successfully!")
```

```python
df = pd.read_csv("C:\\Users\\Administrator\\Downloads\\Code Alpha\\Data Analyst Intern\\titanic.csv")

# Fill missing age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

print("✅ Dataset loaded & cleaned!")
df.head()
```

```python
sns.set_style("whitegrid")
sns.set_palette("Set2")
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 11
print("✅ Style set!")
```

```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(x='Survived', data=df, palette=['#FF6B6B', '#6BCB77'])

# Add value labels on bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=13, fontweight='bold')

plt.title("🚢 Survival Count — Titanic", fontsize=14, fontweight='bold')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'], fontsize=12)
plt.ylabel("Number of Passengers")
plt.savefig("viz1_survival_count.png", bbox_inches='tight')
plt.show()
print("✅ Chart 1 saved!")
```

```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(x='Sex', hue='Survived', data=df,
                   palette=['#FF6B6B', '#6BCB77'])

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11)

plt.title("👫 Survival by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(['Did Not Survive', 'Survived'])
plt.savefig("viz2_survival_gender.png", bbox_inches='tight')
plt.show()
print("✅ Chart 2 saved!")
```

```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(x='Pclass', hue='Survived', data=df,
                   palette=['#FF6B6B', '#6BCB77'])

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11)

plt.title("🎫 Survival by Passenger Class", fontsize=14, fontweight='bold')
plt.xlabel("Class (1=First  2=Second  3=Third)")
plt.ylabel("Number of Passengers")
plt.legend(['Did Not Survive', 'Survived'])
plt.savefig("viz3_survival_class.png", bbox_inches='tight')
plt.show()
print("✅ Chart 3 saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Survived', bins=30,
             palette=['#FF6B6B', '#6BCB77'], alpha=0.7)

plt.title("🎂 Age Distribution by Survival", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(['Did Not Survive', 'Survived'])
plt.savefig("viz4_age_distribution.png", bbox_inches='tight')
plt.show()
print("✅ Chart 4 saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df,
            palette=['#FF6B6B', '#6BCB77'])

plt.title("💰 Fare Distribution by Class & Survival", fontsize=14, fontweight='bold')
plt.xlabel("Class (1=First  2=Second  3=Third)")
plt.ylabel("Fare (£)")
plt.legend(['Did Not Survive', 'Survived'])
plt.savefig("viz5_fare_boxplot.png", bbox_inches='tight')
plt.show()
print("✅ Chart 5 saved!")
```

```python
plt.figure(figsize=(6, 6))
survival_counts = df['Survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
colors = ['#FF6B6B', '#6BCB77']
explode = (0.05, 0.05)

plt.pie(survival_counts, labels=labels, colors=colors,
        autopct='%1.1f%%', explode=explode,
        shadow=True, startangle=140,
        textprops={'fontsize': 13})

plt.title("🥧 Survival Rate — Titanic", fontsize=14, fontweight='bold')
plt.savefig("viz6_survival_pie.png", bbox_inches='tight')
plt.show()
print("✅ Chart 6 saved!")
```

```python
plt.figure(figsize=(9, 6))

# Select only numeric columns
numeric_df = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]

sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f",
            cmap='RdYlGn', linewidths=0.5,
            square=True, cbar_kws={"shrink": 0.8})

plt.title("🔥 Correlation Heatmap", fontsize=14, fontweight='bold')
plt.savefig("viz7_correlation_heatmap.png", bbox_inches='tight')
plt.show()
print("✅ Chart 7 saved!")
```

```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(x='Embarked', hue='Survived', data=df,
                   palette=['#FF6B6B', '#6BCB77'])

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11)

plt.title("🚢 Survival by Embarkation Point", fontsize=14, fontweight='bold')
plt.xlabel("Port (C=Cherbourg  Q=Queenstown  S=Southampton)")
plt.ylabel("Count")
plt.legend(['Did Not Survive', 'Survived'])
plt.savefig("viz8_embarkation.png", bbox_inches='tight')
plt.show()
print("✅ Chart 8 saved!")
```

```python
fig = px.scatter(df, x='Age', y='Fare',
                 color=df['Survived'].map({0: 'Did Not Survive', 1: 'Survived'}),
                 symbol='Sex',
                 title='🔵 Age vs Fare — Interactive Chart',
                 color_discrete_map={
                     'Did Not Survive': '#FF6B6B',
                     'Survived': '#6BCB77'
                 },
                 labels={'color': 'Survival Status'})

fig.update_layout(title_font_size=16)
fig.show()
print("✅ Interactive chart displayed!")
```
